```python
"""
Simple ADMET Properties Prediction Example using RDKit
Predicts basic ADMET-related descriptors for several compounds.
These descriptors serve as proxies for ADMET properties:
- Molecular Weight (MW): Related to absorption and distribution
- LogP: Octanol-water partition coefficient (absorption, distribution)
- TPSA: Topological Polar Surface Area (absorption)
- HBD: Hydrogen Bond Donors (absorption)
- HBA: Hydrogen Bond Acceptors (absorption)
- Rotatable Bonds: Related to metabolism and flexibility
- QED: Quantitative Estimate of Drug-likeness (overall drug potential, indirect toxicity proxy)

Note: This is a basic computational prediction using descriptors.
For more accurate ADMET, consider ML models or tools like SwissADME.
"""

import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors, QED

# List of sample compounds (SMILES strings)
# Examples: HER2 inhibitors and common drugs
smiles_list = [
    'CS(=O)(=O)CCNCC1=CC=C(O1)C2=CC3=C(C=C2)N=CN=C3NC4=CC(=C(C=C4)OCC5=CC(=CC=C5)F)Cl',  # Lapatinib
    'CCOC1=C(C=C2C(=C1)N=CC(=C2NC3=CC(=C(C=C3)OCC4=CC=CC=N4)Cl)C#N)NC(=O)C=CCN(C)C',  # Neratinib
    'CC1=C(C=CC(=C1)NC2=NC=NC3=C2C=C(C=C3)NC4=NC(CO4)(C)C)OC5=CC6=NC=NN6C=C5',  # Tucatinib
    'CC(=O)OC1=CC=CC=C1C(=O)O',  # Aspirin
    'CC(C)CC1=CC=C(C=C1)C(C)C(=O)O',  # Ibuprofen
    'CN1C=NC2=C1C(=O)N(C(=O)N2C)C',  # Caffeine
]
```

```python
# Function to compute ADMET-related properties
def compute_admet_properties(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None

    properties = {
        'SMILES': smiles,
        'Molecular Weight': Descriptors.MolWt(mol),
        'LogP': Descriptors.MolLogP(mol),
        'TPSA': Descriptors.TPSA(mol),
        'HBD': Descriptors.NumHDonors(mol),
        'HBA': Descriptors.NumHAcceptors(mol),
        'Rotatable Bonds': Descriptors.NumRotatableBonds(mol),
        'QED': QED.qed(mol)  # Drug-likeness score (0-1, higher is better)
    }

    return properties
```

```python
# Compute properties for all compounds
results = []
for smiles in smiles_list:
    props = compute_admet_properties(smiles)
    if props:
        results.append(props)

# Create DataFrame for tabular output
df = pd.DataFrame(results)

# Display results
print("ADMET Properties Prediction Results:")
print(df.to_string(index=False))

# Simple interpretation
print("\nInterpretation Guidelines:")
print("- Absorption: Good if LogP 0-3, TPSA <140, HBD <=5, HBA <=10, MW <500")
print("- Distribution: LogP >0 for better BBB penetration")
print("- Metabolism: Fewer rotatable bonds may indicate stability")
print("- Excretion: Related to MW and LogP")
print("- Toxicity: Higher QED suggests lower toxicity risk (proxy)")

# Export to CSV (optional)
df.to_csv('admet_predictions.csv', index=False)
print("\nResults exported to 'admet_predictions.csv'")
```
