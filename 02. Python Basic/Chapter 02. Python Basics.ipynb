A variable in Python is simply a name that refers to a value. To assign a value to a variable, use the assignment operator "=". For example:

**(2) Strings (str): The Foundation of Chemical Representations**

**Characteristics of Strings**

A string (str) in Python is an immutable sequence of characters. In cheminformatics, strings are fundamental because they store molecular representations such as:

- Molecular formulas: "C6H6" for benzene.
- SMILES notation: "C1=CC=CC=C1" for benzene.
- InChI keys: "InChI=1S/C6H6/c1-2-4-6-5-3-1/h1-6H".

**String Manipulation in Cheminformatics**

Python provides a variety of string formatting operations, which are used as follows:

**Percent (%) Formatting**

This formatting is an older method reminiscent of C-style printf formatting. It uses format specifiers within a string to denote where variables should be inserted.

Example:

```python
# Chemical compound information
compound_name = "Water"
molecular_formula = "H2O"
molar_mass = 18.015  # grams per mole

# Using percent formatting
formatted_string = "Compound: %s, Formula: %s, Molar Mass: %.3f g/mol" % (compound_name, molecular_formula, molar_mass)
print(formatted_string)
# Output: Compound: Water, Formula: H2O, Molar Mass: 18.015 g/mol
```

Explanation:

- %s is used for string substitution.
- %.3f specifies a floating-point number with three decimal places.

**str.format() Method**

Introduced in Python 3, this method offers more flexibility and readability compared to percent formatting.

Example:

```python
# Chemical reaction information
reactant1 = "Hydrogen"
reactant2 = "Oxygen"
product = "Water"
reaction_condition = "Ignition"

# Using str.format()
formatted_string = "When {} and {} react under {}, they produce {}.".format(reactant1, reactant2, reaction_condition, product)
print(formatted_string)
# Output: When Hydrogen and Oxygen react under Ignition, they produce Water.
```

Explanation:

- Curly braces {} are placeholders for variables.
- The format() method replaces these placeholders with the provided arguments in order.

**f-strings (Formatted String Literals)**

Introduced in Python 3.6, f-strings provide a concise and intuitive way to embed expressions inside string literals.

```python
# Concentration calculations
initial_concentration = 0.1  # Molar (M)
dilution_factor = 10

# Using f-string
final_concentration = initial_concentration / dilution_factor
formatted_string = f"After dilution, the concentration changes from {initial_concentration:.2f} M to {final_concentration:.3f} M."
print(formatted_string)
# Output: After dilution, the concentration changes from 0.10 M to 0.010 M.
```

Explanation:

- The f prefix indicates a formatted string literal.
- Expressions inside curly braces {} are evaluated at runtime.
- :.2f and :.3f specify the number of decimal places for floating-point numbers.

**Template Strings (string.Template)**

This method is useful for scenarios requiring safer substitution, such as when inputs are not fully controlled, to mitigate code injection risks.

Example:

```python
from string import Template

# Chemical element information
template = Template("Element: $name, Atomic Number: $atomic_number, Symbol: $symbol")

# Using Template
formatted_string = template.substitute(name="Sodium", atomic_number=11, symbol="Na")
print(formatted_string)
# Output: Element: Sodium, Atomic Number: 11, Symbol: Na
```

Explanation:

- Template objects use $ to denote placeholders.
- The substitute() method replaces placeholders with actual values.

### 2.4.1 Lists

In Python, a **list** is an ordered collection of items, which can be of any data type, including numbers, strings, or even other lists. Lists are one of the most commonly used data structures in Python due to their flexibility, ease of use, and ability to store items of different types. This section will dive deeper into how lists work, how to manipulate them and provide chemistry-related examples to illustrate their utility.

**(1) Creating and Accessing Lists**

A list in Python is defined by placing items (elements) inside square brackets "\[\]" and separating them with commas. You can store elements of any data type in a list, and they can be of mixed types.

Example:

```python
chemical_elements = ["Hydrogen", "Helium", "Lithium", "Beryllium"]
```

To access an element in a list, you use indexing. Python indexing is zero-based, meaning the first item in the list is at index 0, the second at index 1, and so on.

Example:

```python
# Accessing elements
first_element = chemical_elements[0]  # Hydrogen
last_element = chemical_elements[-1]  # Beryllium (negative index for reverse access)

print(first_element)  # Outputs: Hydrogen
print(last_element)   # Outputs: Beryllium
```

**(2) Modifying Lists**

Lists are **mutable**, meaning their contents can be changed after creation. You can modify, add, or remove elements from a list.

Modifying an Element:

```python
chemical_elements[2] = "Lithium (Li)"  # Change 'Lithium' to 'Lithium (Li)'
print(chemical_elements)  
# Outputs: ['Hydrogen', 'Helium', 'Lithium (Li)', 'Beryllium']
```

**Appending an Element:** To add an element to the end of a list, use the append() method:

```python
chemical_elements.append("Boron")
print(chemical_elements)  
# Outputs: ['Hydrogen', 'Helium', 'Lithium (Li)', 'Beryllium', 'Boron']
```

**Inserting an Element:** To insert an element at a specific position, use the insert() method:

```python
chemical_elements.insert(2, "Carbon")  # Insert 'Carbon' at index 2
print(chemical_elements)  
# Outputs: ['Hydrogen', 'Helium', 'Carbon', 'Lithium (Li)', 'Beryllium', 'Boron']
```

**Removing an Element:** Use the remove() method to remove the first occurrence of a value:

```python
chemical_elements.remove("Boron")
print(chemical_elements)  # Outputs: ['Hydrogen', 'Helium', 'Carbon', 'Lithium (Li)', 'Beryllium']
```

**Popping an Element:** The pop() method removes an element at a specified index and returns it. If no index is provided, it removes and returns the last element:

```python
popped_element = chemical_elements.pop(3)  # Remove element at index 3 (Lithium (Li))
print(popped_element)  # Outputs: Lithium (Li)
print(chemical_elements)  # Outputs: ['Hydrogen', 'Helium', 'Carbon', 'Beryllium']
```

**(3) Nested Lists**

Python allows lists to contain other lists, making them ideal for representing matrices or multi-dimensional data. A nested list is simply a list that contains other lists as its elements.

Example (representing a 2D array for a periodic table of elements with atomic numbers):

```python
periodic_table = [
    ["H", "He", "Li", "Be"],
    ["B", "C", "N", "O"],
    ["F", "Ne", "Na", "Mg"]
]

# Accessing an element in a nested list
element = periodic_table[1][2]  # 'N' from the second row, third column
print(element)  # Outputs: N
```

You can perform similar operations on nested lists as you would with regular lists, but be mindful of the index levels when accessing or modifying nested items.

**(4) List Slicing**

You can also extract a portion (or slice) of a list using the slicing syntax list\[start:end\], where start is the index of the first element to include, and end is the index of the element just past the last element to include.

Example:


```python
elements = ["Hydrogen", "Helium", "Lithium", "Beryllium", "Boron"]
sublist = elements[1:4]  # Slice from index 1 to 3 (Helium, Lithium, Beryllium)
print(sublist)  # Outputs: ['Helium', 'Lithium', 'Beryllium']
```


You can also use negative indices for slicing from the end of the list:


```python
last_two_elements = elements[-2:]  # Last two elements: ['Beryllium', 'Boron']
print(last_two_elements)
```

**(5) List Comprehensions**

List comprehensions provide a concise way to create lists. A list comprehension consists of an expression followed by a for loop, optionally, an if condition. They allow for more compact and readable code than using loops.

Syntax:

```python
[expression for item in iterable if condition]
```

Example: Creating a list of atomic numbers from the names of elements:

```python
elements = ["Hydrogen", "Helium", "Lithium", "Beryllium", "Boron"]
atomic_numbers = [i + 1 for i in range(len(elements))]  # Creating atomic numbers from 1 to N
print(atomic_numbers)  # Outputs: [1, 2, 3, 4, 5]
```

In this example:

- i + 1 generates the atomic numbers.
- range(len(elements)) loops over the indices of the elements list.

Example with a condition: Filtering out elements that have more than six letters:


```python
long_elements = [element for element in elements if len(element) > 6]
print(long_elements)  # Outputs: ['Lithium', 'Beryllium']
```

List comprehensions are particularly useful for data transformation and filtering tasks in a concise manner.

**(6) Common List Methods**

extend(): Adds all elements from an iterable (like another list) to the end of the list:

```python
elements = ["Hydrogen", "Helium"]
elements.extend(["Lithium", "Beryllium"])
print(elements)  # Outputs: ['Hydrogen', 'Helium', 'Lithium', 'Beryllium']
```

sort(): Sorts the list in place. To sort in reverse order, use sort(reverse=True):

```python
elements = ["Helium", "Hydrogen", "Lithium", "Beryllium"]
elements.sort()
print(elements)  # Outputs: ['Beryllium', 'Helium', 'Hydrogen', 'Lithium']
```

reverse(): Reverses the elements of the list in place:

```python
elements.reverse()
print(elements)  # Outputs: ['Lithium', 'Hydrogen', 'Helium', 'Beryllium']
```

count(): Counts how many times a specific element appears in the list:

```python
elements = ["Helium", "Hydrogen", "Hydrogen"]
print(elements.count("Hydrogen"))  # Outputs: 2
```

index(): Returns the index of the first occurrence of a specified value:

```python
print(elements.index("Hydrogen"))  # Outputs: 1
```

**(7) Cases in Chemistry**

Here are some examples of how lists can be used in chemistry-related Python programs:

**Molar Mass Calculation**: Calculate the molar mass of a compound by summing the atomic masses of its elements.

```python
atomic_masses = {"H": 1.008, "He": 4.0026, "Li": 6.94, "Be": 9.0122, "O": 15.999}
compound = ["H", "O", "H"]
molar_mass = sum(atomic_masses[element] for element in compound)
print(f"Molar mass: {molar_mass} g/mol")
```

**Chemical Reactions**: Represent a simple chemical reaction and calculate the total number of atoms of each element.

```python
reactants = {"H2O": 2, "CO2": 3}
products = {"C6H12O6": 1, "O2": 2}

elements_in_reactants = ["H", "O", "C"] * reactants["H2O"] + ["C", "O"] * reactants["CO2"]
elements_in_products = ["C", "H", "O"] * products["C6H12O6"] + ["O"] * products["O2"]

print("Reactants elements:", elements_in_reactants)
print("Products elements:", elements_in_products)
```

**Periodic Table Operations**: Search and manipulate lists of elements to find properties or categorize them based on certain attributes.

```python
elements = ["Hydrogen", "Helium", "Lithium", "Beryllium"]
metals = ["Lithium", "Beryllium"]

non_metals = [element for element in elements if element not in metals]
print("Non-metals:", non_metals)  # Outputs: ['Hydrogen', 'Helium']
```

### 2.4.3 Sets

A set is an unordered collection of unique items. Sets are useful when you need to store items without worrying about duplicates and when the order of items does not matter.

## 2.5 Control Flow

Control flow in Python refers to the order in which individual statements, instructions, or function calls are executed or evaluated. In this section, we will explore how to control the flow of execution in your Python programs using conditional statements, loops, and other control structures.

### 2.5.1 Conditional Statements

Conditional statements allow you to execute certain blocks of code depending on whether a condition is true or false. The most common conditional statements in Python are if, elif, and else.

**if Statement:** The if statement is used to check the condition. If the condition is True, the block of code inside the if statement is executed.

Example:

```python
x = 10
if x > 5:
    print("x is greater than 5")
# Outputs: x is greater than 5
```

**elif Statement:** The elif (short for “else if”) allows you to check multiple conditions in sequence. If the initial if condition is False, the elif condition will be checked.

Example:

```python
x = 10
if x < 5:
    print("x is less than 5")
elif x == 10:
    print("x is equal to 10")
# Outputs: x is equal to 10
```

**else Statement:** The else block executes if all previous conditions are False. You can use it to provide a default action when no conditions are met.

Example:


```python
x = 2
if x > 5:
    print("x is greater than 5")
elif x == 3:
    print("x is equal to 3")
else:
    print("x is neither greater than 5 nor equal to 3")
# Outputs: x is neither greater than 5 nor equal to 3
```

### 2.5.2 Logical Operators

Logical operators allow you to combine multiple conditions. Python supports the following logical operators:

**and:** Returns True if both conditions are True.

**or:** Returns True if at least one of the conditions is True.

**not:** Reverses the result of the condition.

Examples:


```python
x = 10
y = 5
if x > 5 and y < 10:
    print("Both conditions are True")  # Outputs: Both conditions are True

if x < 5 or y > 0:
    print("At least one condition is True")  # Outputs: At least one condition is True

if not(x == 10):
    print("x is not 10")  # This will not print because x == 10
```

Loops allow you to execute a block of code multiple times. Python supports two main types of loops: the for loop and the while loop.

**for Loop:** A for loop iterates over a sequence (such as a list, tuple, or string) and executes the block of code for each item in the sequence.

Example:


```python
elements = ["H", "C", "O"]
for element in elements:
    print(element)
# Outputs
# H
# C
# O
```


You can also use the function range() to iterate over a sequence of numbers:

```python
for i in range(5):  # Loops from 0 to 4
    print(i)
```

**while Loop:** A while loop executes a block of code as long as a specified condition is True. Be careful with while loops to avoid creating infinite loops.

Example:

```python
x = 0
while x < 5:
    print(x)
    x += 1  # Increment x by 1
# Outputs
# 0
# 1
# 2
# 3
# 4
```

Python provides the following statements to alter the flow of loops:

**break:** The break statement is used to exit a loop prematurely.

Example:

```python
for i in range(10):
    if i == 5:
        break  # Exits the loop when i equals 5
    print(i)
# Outputs
# 0
# 1
# 2
# 3
# 4
```

Example:

```python
for i in range(5):
    if i == 2:
        continue  # Skips the iteration when i equals 2
    print(i)
# Outputs
# 0
# 1
# 3
# 4
```

Example:

```python
for i in range(5):
    pass  # No operation is performed
```

In Python, functions are first-class objects. It means that functions can be passed as arguments to other functions, returned as values from other functions, and assigned to variables.

```python
def greet(name):
    return f"Hello, {name}!"

def process_function(func, name):
    return func(name)

result = process_function(greet, "Alice")
print(result)  # Outputs: Hello, Alice!
```

Example of returning a function from another function:

```python
def outer():
    def inner():
        return "This is the inner function"
    return inner

result = outer()  # outer returns the inner function
print(result())  # Outputs: This is the inner function
```

### 2.6.7 Recursion

A function can call itself. This technique is called **recursion**. Recursive functions are useful for problems that can be broken down into smaller, similar subproblems, such as computing factorials or traversing trees.

The key to writing a recursive function is defining a **base case**, which prevents the function from calling itself indefinitely.

Example (Factorial):

```python
def factorial(n):
    if n == 1:
        return 1  # Base case
    else:
        return n * factorial(n - 1)  # Recursive case

print(factorial(5))  # Outputs: 120
```

## 2.7 Classes

In Python, classes are the foundation of object-oriented programming (OOP). A class is a blueprint for creating objects (instances), which are collections of data and methods that act on that data. By using classes, you can model real-world concepts and organize your code in a more structured way. This section will introduce you to the basics of creating and using classes in Python.

### 2.7.1 Defining a Class

A class in Python is defined through the class keyword followed by the class name. The body of the class contains its attributes (variables) and methods (functions).

Example:

```python
class ChemicalElement:
    # Constructor to initialize the object
    def __init__(self, name, symbol, atomic_weight):
        self.name = name            # Instance variable for element name
        self.symbol = symbol        # Instance variable for element symbol
        self.atomic_weight = atomic_weight  # Instance variable for atomic weight
 
    # Method to display information about the chemical element
    def display_info(self):
        print(f"Element: {self.name} (Symbol: {self.symbol}), Atomic Weight: {self.atomic_weight}")
```

In this example:

- ChemicalElement is the name of the class.
- The \__init__ method is a special method known as the constructor, which is used to initialize the object's attributes when an instance is created.
- The display_info method is an instance method that can be called on any instance of the ChemicalElement class.

### 2.7.2 Creating an Object (Instance)

To create an object (instance) of a class, you can call the class as if it is a function, passing the required arguments to the constructor.

Example:

```python
hydrogen = ChemicalElement("Hydrogen", "H", 1.008)
```

Here, hydrogen is an instance of the ChemicalElement class, with the name attribute set to "Hydrogen", the symbol attribute set to "H", and the atomic_weight attribute set to 1.008.

### 2.7.3 Accessing Attributes and Methods

Once an instance is created, you can access its attributes and methods using the notation ".".

Example:

```python
# Accessing instance attributes
print(hydrogen.name)          # Outputs: Hydrogen
print(hydrogen.atomic_weight) # Outputs: 1.008
 
# Calling a method on the instance
hydrogen.display_info()       # Outputs: Hydrogen (Symbol: H) says: Check out my atomic weight!
```

### 2.7.4 The \__init__ Method (Constructor)

The \__init__method is the constructor in Python. It is automatically called when an object of the class is created. You can use the __init__ method to initialize the object's attributes. This method always takes at least one argument: self, which refers to the current instance of the class.

Example:

```python
class ChemicalCompound:
    def __init__(self, name, formula, molecular_weight):
        self.name = name               # Instance variable for compound name
        self.formula = formula         # Instance variable for chemical formula
        self.molecular_weight = molecular_weight  # Instance variable for molecular weight

    def display_info(self):
        print(f"{self.name} ({self.formula}) - Molecular Weight: {self.molecular_weight}")
        
# Creating an instance of ChemicalCompound
water = ChemicalCompound("Water", "H2O", 18.015)

# Displaying information about the instance
water.display_info()  # Outputs: Water (H2O) - Molecular Weight: 18.015
```

### 2.7.5 Instance Variables and Methods

Instance Variables: These are variables that are tied to a particular instance of the class. They are usually initialized inside the \__init__ method using self.

Example:

```python
class Person:
    def __init__(self, name, age):
        self.name = name  # Instance variable
        self.age = age    # Instance variable
```

Instance Methods: These are functions defined within the class that operate on instance variables. They are called using an instance of the class.

Example:


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
        
person1 = Person("Alice", 25)
person1.greet()  
# Outputs: Hello, my name is Alice and I am 25 years old.
```

### 2.7.6 Class Variables and Methods

Class variables are shared across all instances of a class. They are declared inside the class but outside of any methods. Class methods, on the other hand, are methods that are bound to the class rather than an instance. They take cls as their first argument instead of self.

**Class Variables:** These are variables that are shared across all instances of the class.

Example:

```python
class Element:
    category = "Chemical Element"  # Class variable
    
    def __init__(self, name, symbol, atomic_number):
        self.name = name
        self.symbol = symbol
        self.atomic_number = atomic_number
 
    def display_info(self):
        print(f"Element: {self.name} (Symbol: {self.symbol}, Atomic Number: {self.atomic_number})")
        
element1 = Element("Hydrogen", "H", 1)
element2 = Element("Oxygen", "O", 8)
 
print(element1.category)  # Output: Chemical Element
print(element2.category)  # Output: Chemical Element
```

**Class Methods:** These methods are bound to the class and can be used to modify class variables or perform operations that affect the class as a whole. They receive the class itself as the first argument, conventionally named cls.

Example:

```python
class Element:
    category = "Chemical Element"  # Class variable
    
    def __init__(self, name, symbol, atomic_number):
        self.name = name
        self.symbol = symbol
        self.atomic_number = atomic_number

    @classmethod
    def change_category(cls, new_category):
        cls.category = new_category
        print(f"The category has been updated to: {cls.category}")  # Demonstrating the use of cls

# Changing class variable 'category' using a class method and demonstrating the use of cls
Element.change_category("Fundamental Substance")  # Output: The category has been updated to: Fundamental Substance
print(Element.category)  # Output: Fundamental Substance
```

In the example, the change_category class method uses cls to refer to the class itself, allowing it to alter the class variable category. This modification affects all instances of the class, reflecting the change in the class variable's value regardless of which instance accesses it.

### 2.7.7 Inheritance

Inheritance is a mechanism that allows you to create a new class that is a modified version of an existing class. The new class inherits attributes and methods from the existing class and can also have additional attributes or methods.

Example:

```python
class Substance:
    def __init__(self, name):
        self.name = name
    
    def describe(self):
        print(f"{self.name} is a general substance.")

class Element(Substance):  # Element inherits from Substance
    def __init__(self, name, symbol):
        # Calling the constructor of the base class (Substance)
        super().__init__(name)
        self.symbol = symbol

    def describe(self):  # Overriding the describe method
        print(f"{self.name} (Symbol: {self.symbol}) is a chemical element.")

element = Element("Oxygen", "O")
element.describe()  # Outputs: Oxygen (Symbol: O) is a chemical element.
```

In this example:

- The Element class inherits from the Substance class, so it has access to the name attribute and describe method from Substance.
- The describe method is overridden in the Element class to provide a more specific implementation.

### 2.7.8 Polymorphism

Polymorphism allows objects of different classes to be treated as objects of a common superclass. It also allows methods to have the same name but behave differently based on the object calling them.

Example:

```python
class Compound(Substance):
    def describe(self):
        print(f"{self.name} is a chemical compound.")


substances = [Element("Oxygen", "O"), Compound("Water")]

for substance in substances:
    substance.describe()

# Outputs:
# Oxygen (Symbol: O) is a chemical element.
# Water is a chemical compound.
```

Even though the objects are different types (Element and Compound), they both have a describe method, and Python calls the appropriate method for each object (this is polymorphism in action).

### 2.7.9 Encapsulation

Encapsulation is the practice of restricting direct access to some of an object's attributes or methods and providing access through public methods (getters and setters). It helps protect the internal state of an object and allows for more controlled interactions.

Example:


```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        # Private variable, should not be accessed directly
        self.__balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
    
    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())  # Outputs: 1300
```

In this example, the \__balance variable is private, meaning it cannot be accessed directly from outside the class. Instead, access is provided via methods (deposit, withdraw, and get_balance), which ensure the integrity of the balance.

## 2.8 Modules

```python
# math_operations.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"
```

### 2.8.3 The \__name__ Variable

Each Python module has a built-in variable called \__name_\_. When a module is run directly, the value of \__name__is set to "\__main_\_". When the module is imported into another script, the value of \__name__ is set to the name of the module.

**(3) The math Module**

The math module is indispensable for performing numerical calculations, a frequent requirement in cheminformatics. It provides functions for trigonometric calculations, logarithms, and constants like π and e—useful in molecular modeling, geometry optimizations, and energy calculations.

Example: Calculating Interatomic Distances


```python
import math

def compute_distance(coord1, coord2):
    """Calculate Euclidean distance between two points (atoms)."""
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(coord1, coord2)))

# Example atomic coordinates in 3D space
atom1 = (0.0, 0.0, 0.0)
atom2 = (1.0, 1.0, 1.0)
distance = compute_distance(atom1, atom2)
print("Distance between atoms:", distance)
```

**(6) The re Module**

Regular expressions are powerful for parsing and validating text data. In cheminformatics, they are particularly useful for processing chemical file formats (such as SMILES strings, InChI keys) and extracting molecular properties from textual reports or logs. The re module offers extensive capabilities for pattern matching and text manipulation.

Example: Validating a SMILES String

```python
import re

# A simple example SMILES string for benzene
smiles = "C1=CC=CC=C1"
# Regular expression pattern for a basic validation of SMILES characters
pattern = r'^[BCNOFPSIbcnofpsi0-9@+\-\[\]\(\)=#$]+$'

if re.match(pattern, smiles):
    print("Valid SMILES string format.")
else:
    print("Invalid SMILES string format.")
```

This example demonstrates how regular expressions can enforce expected patterns in chemical data formats, ensuring that the data conforms to standards before further processing.

This process is essential for saving chemical experiment results in a standardized format that can be shared across platforms or uploaded to chemical databases.

## 2.13 Exceptions

Exceptions are a way of handling errors in Python. Instead of having your program crash when an error occurs, exceptions allow you to catch and handle errors gracefully. This section will introduce the basics of exception handling, including how to raise and catch exceptions and how to manage errors effectively in your Python programs.

### 2.13.1 What Are Exceptions

An exception is the runtime error that occurs during the execution of a program. When an error occurs, Python raises an exception, which can interrupt the flow of the program. If the exception is not handled, the program will stop execution.

For example:

```python
>>>13 * (1/0)
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
    13 * (1/0)
          ~^~
ZeroDivisionError: division by zero
>>>3 + chem*2
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
    3 + chem*2
        ^^^^
NameError: name 'chem' is not defined
>>>'1' + 2
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
    '1' + 2
    ~~~~^~~
TypeError: can only concatenate str (not "int") to str
```

### 2.13.2 Handling Exceptions with try and except

Python provides a try and except block to handle exceptions. Code that may raise an exception is placed in the try block, and if an exception occurs, the except block handles it.

Example:


```python
try:
    x = 10
    y = 0
    z = x / y  # This raises ZeroDivisionError
except ZeroDivisionError:
    print("Cannot divide by zero!")
#Outputs: Cannot divide by zero!
```

### 2.13.3 Catching Multiple Exceptions

You can catch multiple types of exceptions using multiple except blocks or by specifying multiple exceptions in a single block.

Example:

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    result = x / y
except ZeroDivisionError:
    print("You can't divide by zero!")
except ValueError:
    print("Invalid input. Please enter numbers.")
```

### 2.13.4 The else and finally Blocks

- **else:** The else block is executed if no exceptions are raised in the try block.
- **finally:** The finally block is executed no matter what, whether an exception is raised or not. It’s typically used for cleanup actions.

Example:

```python
try:
    x = 10
    y = 5
    result = x / y
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"The result is {result}")
finally:
    print("Execution completed.")

# Outputs
# The result is 2.0
# Execution completed.
```

### 2.13.5 Raising Exceptions

You can raise exceptions manually using the raise keyword. It is useful when you want to signal an error condition that your program should handle.

Example:

```python
def divide(x, y):
    if y == 0:
        raise ValueError("Cannot divide by zero!")
    return x / y

try:
    result = divide(10, 0)
except ValueError as e:
    print(e)  # Outputs: Cannot divide by zero!
```

### 2.13.6 Custom Exceptions

You can define your own exceptions by creating a class that inherits from the built-in Exception class.

Example:

```python
class NegativeNumberError(Exception):
    pass

def check_positive(x):
    if x < 0:
        raise NegativeNumberError("Negative numbers are not allowed!")
    return x

try:
    check_positive(-5)
except NegativeNumberError as e:
    print(e)  # Outputs: Negative numbers are not allowed!
```

### 2.14.2 Writing Simple Tests with unittest

Python’s unittest module is a built-in framework for writing and running tests. To write a test, you need to create a class that inherits from unittest. TestCase and define methods that test specific functionality. Each test method must start with the word test, as shown in Figure 2.11:

<img src="./Figure2.11.png" width=1000 height=800>

\*\*\* Insert FIGURE 2.11 HERE \*\*\*

Caption: Writing Simple Tests with unittest.

```python
import unittest

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

class TestMathOperations(unittest.TestCase):
    
    def test_add(self):
        self.assertEqual(add(2, 3), 5)
        self.assertEqual(add(-1, 1), 0)
        
    def test_subtract(self):
        self.assertEqual(subtract(5, 3), 2)
        self.assertEqual(subtract(-1, 1), -2)

if __name__ == "__main__":
    unittest.main()
```

In this example:

- test_add() checks that the add function works correctly for both positive and negative numbers.
- test_subtract() checks that the subtract function works as expected.

### 2.14.3 Assertions in Tests

Assertions are used in unittest to check if the results of your functions match the expected output. Common assertion methods include:

- assertEqual(a, b): Checks if a is equal to b.
- assertNotEqual(a, b): Checks if a is not equal to b.
- assertTrue(x): Checks if x is True.
- assertFalse(x): Checks if x is False.
- assertRaises(exception, callable, \*args, \*\*kwargs): Checks if a specific exception is raised when calling a function.

Example:


```python
def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b

class TestDivision(unittest.TestCase):
    
    def test_divide(self):
        self.assertEqual(divide(10, 2), 5)
        self.assertRaises(ValueError, divide, 10, 0)

if __name__ == "__main__":
    unittest.main()
```

### 2.15.5 Logging Instead of Printing

For larger applications, you may want to use logging rather than print statements. The logging module allows you to output diagnostic information, track errors, and set different levels of severity (e.g., DEBUG, INFO, WARNING, ERROR, and CRITICAL).

Example:


```python
import logging

logging.basicConfig(level=logging.DEBUG)

def divide(x, y):
    logging.debug(f"Dividing {x} by {y}")
    if y == 0:
        logging.error("Attempt to divide by zero!")
        return None
    return x / y

result = divide(10, 2)
```

Using logging allows for better control over output and makes it easier to disable or redirect log messages as needed.

## 2.16 Profiling

Profiling is the process of measuring the performance of your Python code to identify bottlenecks and optimize efficiency. Python provides tools for profiling the execution time of your code and understanding how much time is spent in different parts of the program.

### 2.16.1 Why Profile Your Code

Profiling is important for:

- **Optimizing performance:** Identifying inefficient code and improving it to make your program faster.
- **Understanding resource usage:** Understanding where your program uses the most memory or CPU.
- **Focusing efforts:** By knowing where the bottlenecks are, you can focus your optimization efforts on the areas that matter most.

### 2.16.2 Using the cProfile Module

Python's built-in cProfile module is used to profile Python programs. It provides detailed information about the time spent in each function, how many times each function is called, and how much time is spent in each call.

Example:

```python
import cProfile

def my_function():
    total = 0
    for i in range(1000000):
        total += i
    return total

cProfile.run('my_function()')
```

- ncalls: The number of calls to the function.
- tottime: The total time spent in the function (excluding time spent in sub-functions).
- cumtime: The cumulative time spent in the function and all sub-functions.
- percall: The average time per call.

You can also save the output of cProfile to a file for later analysis:

```python
cProfile.run('my_function()', 'output.prof')
```

### 2.16.3 Visualizing Profiling Data with pstats and gprof2dot

You can use the pstats module to read and analyze the profiling results saved to a file. You can also visualize the profiling data using tools like gprof2dot to generate a call graph.

Example of using pstats:


```python
import pstats

profile = pstats.Stats('output.prof')
profile.sort_stats('cumulative').print_stats()
```

o visualize the results using gprof2dot, install it via pip:

```python
pip install gprof2dot
```

Then, use it to generate a call graph:

```python
gprof2dot -f pstats output.prof | dot -Tpng -o output.png
```

### 2.16.4 Using time for Quick Timing

For simple performance measurements, you can use the time module to record the execution time of specific code blocks.

Example:

```python
import time

start_time = time.time()

# Code to be timed
my_function()

end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```
