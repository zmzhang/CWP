### Import necessary libraries

```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import py3Dmol
```

### Create molecule

```python
smi = 'O=C(O)c1ccccc1C(=O)O' # Aspirin
mol = Chem.MolFromSmiles(smi)
```

### Display 2D structure

```python
Draw.MolToImage(mol)
```

### Generate 3D coordinates

```python
mol_with_hydrogens = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol_with_hydrogens)
AllChem.MMFFOptimizeMolecule(mol_with_hydrogens)
```

### Display interactive 3D structure

```python
p = py3Dmol.view(width=500, height=400)
p.addModel(Chem.MolToMolBlock(mol_with_hydrogens), 'mol')
p.setStyle({'stick':{}})
p.zoomTo()
p.show()
```
