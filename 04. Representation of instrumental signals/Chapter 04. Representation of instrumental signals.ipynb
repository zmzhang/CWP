#### NMR，Bruker file read example. It should be noted that the code here includes functions for processing the original data

##### import required libraries

```python
import nmrglue as ng
import numpy as np
import os
import pybaselines
from pybaselines import utils
```

##### In this data type, data is saved in folders. Here we first decompress the files.

##### import required libraries

```python
import zipfile
import os
```

##### Extract all files in the ZIP file to the current working directory

```python
# Unzip the archive
zip_file_path = "Chapter_4_data/PG.zip"  # Path to the zip file
extract_path = os.path.join(os.getcwd(), "PG_extracted")  # Target path for extraction

# Extract files
with zipfile.ZipFile(zip_file_path, "r") as files:
    files.extractall(extract_path)

# Ensure the extracted path is correct
nmr_path = os.path.join(extract_path, "PG")  # Assume the extracted folder is named "PG"
```

##### Define the read function

```python
def read_bruker_h_base_pic(nmr_path, bRaw=False, bMinMaxScale=True):  
    nmr_path = os.path.normpath(nmr_path)
    if bRaw:
        dic, fid = ng.fileio.bruker.read(f'{nmr_path}/1')
        zero_fill_size = dic['acqus']['TD']
        fid = ng.bruker.remove_digital_filter(dic, fid)
        fid = ng.proc_base.zf_size(fid, zero_fill_size)
        fid = ng.proc_base.fft(fid)


    else:
        dic, fid = ng.fileio.bruker.read_pdata(f'{nmr_path}/1/pdata/1')
        zero_fill_size = dic['acqus']['TD']
        offset = (float(dic['acqus']['SW']) / 2) - (float(dic['acqus']['O1']) / float(dic['acqus']['BF1']))
        start = float(dic['acqus']['SW']) - offset
        end = -offset
        step = float(dic['acqus']['SW']) / zero_fill_size
        ppms = np.arange(start, end, -step)[:zero_fill_size]

        # Baseline correction
        baseline = pybaselines.whittaker.arpls(fid, lam=100000.0, diff_order=2, max_iter=50, tol=0.001, weights=None)[0]
        fid = fid - baseline

    # normalization
    if bMinMaxScale:
        fid = fid / np.max(fid)
    return {'name': nmr_path.split(os.sep)[-1], 'ppm': ppms, 'fid': fid, 'bRaw': bRaw}
```

##### file read

```python
spectrum_data = read_bruker_h_base_pic(nmr_path, False, True)
```

##### Extract the ppm and fid

```python
ppm = spectrum_data['ppm']
fid = spectrum_data['fid']
```

##### fid is the vector representation of the spectrum, which can be directly used for subsequent processing. We can print its content, shape, and data type to view it.

```python
print("fid:",fid)
print("fid shape:", fid.shape)
print("Data type of fid:", type(fid))
```

##### You can see that it is a vector with 65536 elements, using the numpy.ndarray representation of the form

##### Now plot the spectrum to show

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
plt.plot(ppm, fid, color="black")
plt.xlabel("ppm", fontsize=14, fontweight='bold')
plt.ylabel("fid", fontsize=14, fontweight='bold')
# plt.title("NMR Spectrum-Propylene glycol", fontsize=24, fontweight='bold')
plt.tight_layout()
plt.show()
```

#### GC, netCDF file for GC-MS, read GC only

#### Here we read the data and draw chromatograms for the two-dimensional data and chromatographic dimension of GC-MS data to be introduced later, in order to meet the needs of some users to explore and utilize the separate chromatographic dimension of chromatographic coupling data.

##### import required libraries

```python
import scipy.io.netcdf as nc
import numpy as np
```

##### Difine the CDF file read function

```python
class netcdf_reader:
    def __init__(self, fname, bmmap=True):
        self.filename = fname
        self.f = nc.netcdf_file(fname, 'r', mmap=bmmap)
        self.total_intensity = self.f.variables['total_intensity']
        self.scan_acquisition_time = self.f.variables['scan_acquisition_time']

    def tic(self):
        tic_dict = {'rt': self.scan_acquisition_time.data / 60.0, 'val': self.total_intensity.data}
        return tic_dict
```

##### The tic function is the total ion chromatogram extraction, which means extracting the chromatographic data of all mass-to-charge ratio channels.

##### Difine the plot function. This is another plot form.

```python
from matplotlib.pyplot import figure, plot, xlabel, ylabel, title, show
def plot_tic(tic_dict):
    figure()
    plot(tic_dict['rt'], tic_dict['val'], color='black')
    xlabel('Retention Time (min)', fontsize=14, fontweight='bold')
    ylabel('Intensity', fontsize=14, fontweight='bold')
    # title('Gas Chromatogram-Total Ion Chromatogram (TIC)', fontsize=24, fontweight='bold')
    show()
```

##### Name the save path of the spectrum file as a variable. Note that the CDF data here is the raw GC-MS data, and we are only processing the chromatographic part.

```python
filename = r'Chapter_4_data/Chromatography/zhi10-5vs1.CDF'
```

##### Use the read function to read and extract its total ion chromatogram.

```python
ncr = netcdf_reader(filename, bmmap=False)
tic = ncr.tic()
rt = tic['rt']
val = tic['val']
```

##### val is the vector representation of the GC, which can be directly used for subsequent processing. We can print its content, shape, and data type to view it.

```python
print("val:",val)
print("val shape:", val.shape)
print("Data type of val:", type(val))
```

##### You can see that it is a vector with 3600 elements, using the numpy.ndarray representation of the form

##### Now plot the TIC

```python
plot_tic(tic)
```

### 4.1.3 Electrochemical Signals

#### Voltammetry, txt file read example

##### We use direct reading without library dependencies

##### Name the save path of the spectrum file as a variable

```python
file_path = r"Chapter_4_data/Electrochemical Signals/P1-re-1.4ev-0v.txt"
```

##### Define empty lists to store data

```python
potential = []  # Potential (V)
current = []  # Current (A)
```

##### Directly read the file, parse the data, and find the location of the data.

```python
with open(file_path, 'r') as file:
    data_started = False
    for line in file:
        line = line.strip()
        # Check if the data section has started
        if line.startswith("Potential/V, Current/A"):
            data_started = True
            continue
        if data_started and line:  # Read the data section
            try:
                values = line.split(",")
                potential.append(float(values[0].strip()))
                current.append(float(values[1].strip()))
            except (ValueError, IndexError):
                # Skip lines that cannot be parsed
                continue
```

##### current is the vector representation of the Voltammetry, which can be directly used for subsequent processing. We can print its content, shape, and data type to view it.

```python
print("the first elements of current:",current[:5])
print("current shape:", len(current))
print("Data type of current:", type(current))
```

##### You can see that it is a vector with 4200 elements, using list representation of the form. It should be noted that although a one-dimensional list can represent a vector, it does not have the mathematical operation characteristics of a vector (such as dot product, normalization, etc.). Therefore, data conversion is required, such as the format of numpy.ndarray.

```python
import numpy as np
current = np.array(current)
print("Data type of current:", type(current))
```

##### Now plot the Voltammetry

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(potential, current, color="black", linewidth=1.2)
plt.xlabel("Potential (V)", fontsize=14, fontweight='bold')  # X-axis label
plt.ylabel("Current (A)", fontsize=14, fontweight='bold')  # Y-axis label
# plt.title("Cyclic Voltammetry Curve", fontsize=24, fontweight='bold')  # Plot title
plt.tight_layout()
plt.show()
```

## 4.2 Matrix Representation

### 4.2.1 Data of Hyphenated Instruments

#### GC-MS, CDF file read example

##### import required libraries

```python
import scipy.io.netcdf as nc
import numpy as np
import sys
import matplotlib.pyplot as plt
```

##### Difine the CDF file read function

```python
class netcdf_reader:
    def __init__(self, fname, bmmap=True):
        self.filename = fname
        self.f = nc.netcdf_file(fname, 'r', mmap=bmmap)
        self.mass_values = self.f.variables['mass_values']
        self.intensity_values = self.f.variables['intensity_values']
        self.scan_index = self.f.variables['scan_index']
        self.total_intensity = self.f.variables['total_intensity']
        self.scan_acquisition_time = self.f.variables['scan_acquisition_time']
        self.mass_max = np.max(self.f.variables['mass_range_max'].data)
        self.mass_min = np.max(self.f.variables['mass_range_min'].data)
        if sys.byteorder == "little":
            self.nbo = "<"
        else:
            self.nbo = ">"

    def mat(self, imin, imax, bin):
        f = nc.netcdf_file(self.filename, 'r', mmap=False)
        t = f.variables['scan_acquisition_time'].data / 60.0
        mass_values = f.variables['mass_values']

        intensity_values = f.variables['intensity_values']

        scan_index = f.variables['scan_index']
        scan_index_end = np.hstack((scan_index.data, np.array([len(intensity_values.data)], dtype=int)))
        mass_max = np.max(f.variables['mass_range_max'].data)

        mass_min = np.max(f.variables['mass_range_min'].data)

        mz = np.linspace(mass_min, mass_max, num=int((mass_max - mass_min) / bin + 1))
        if imin < 0 and imax > len(t):
            print('please print suitable index of retention time')
            exit()
        rt = t[imin:imax + 1]
        c = int(imax - imin + 1)
        r = int((mass_max - mass_min) / bin + 1)
        mo = np.zeros((c, r))
        for j in range(imin, imax + 1):
            msnext = scan_index_end[j + 1]
            if msnext > np.shape(mass_values):
                mo[j:imax, :] = []
                break
            else:
                mz_val = mass_values[scan_index_end[j]:scan_index_end[j + 1]]
                sp = intensity_values[scan_index_end[j]:scan_index_end[j + 1]]
                ind = np.round((mz_val - mass_min) / bin)
                position = np.nonzero(mz_val > np.max(mz))
                ind = np.delete(ind, position)
                sp = np.delete(sp, position)
                ind2 = np.unique(ind)
                if np.shape(ind2) != np.shape(ind):
                    sp2 = np.zeros(ind2.shape[0])
                    for i in range(0, ind2.shape[0] - 1):
                        tempind = np.nonzero(ind == ind2[i])
                        sp2[i] = np.sum(sp[tempind[0]])
                else:
                    sp2 = sp
                    ind2 = ind
                mo[j - imin, np.array(ind2, dtype=np.int32)] = sp2
        return {'d': mo, 'rt': rt, 'mz': mz}
```

##### The mat function returns a dictionary with the following key values: d: 2D matrix with shape (imax - imin + 1, (mass_max - mass_min) / bin + 1). rt: array of scan retention time points. mz: array of mass-to-charge ratio range, indicating the mass range after binning. Where imin indicates the index of the start scan time point, imax indicates the index of the end scan time point, and bin indicates the bin width of the mass-to-charge ratio range (m/z).

#### LC-MS, MZML file read example

##### import required libraries

```python
from pyteomics import mzml
import matplotlib.pyplot as plt
import numpy as np
```

##### Name the save path of the spectrum file as a variable. And define empty lists to store data.

```python
mzml_path = r"Chapter_4_data/Chromatography/04-LCMS.mzML"

mz_bin_width = 1  # m/z bin width, adjustable
all_mz = []
all_intensity = []
RT = []
```

##### Collect m/z and intensity from all 

```python
with mzml.MzML(mzml_path) as reader:
    for i, spectrum in enumerate(reader):
        mz_array = spectrum['m/z array']
        intensity_array = spectrum['intensity array']
        rt = spectrum.get('scanList', {}).get('scan', [{}])[0].get('scan start time', i)
        RT.append(rt)
        all_mz.append(mz_array)
        all_intensity.append(intensity_array)
```

##### Merge all m/z to determine bin intervals

```python
mz_min = min([mz.min() for mz in all_mz])
mz_max = max([mz.max() for mz in all_mz])
bins = np.arange(mz_min, mz_max + mz_bin_width, mz_bin_width)
bin_centers = bins[:-1] + mz_bin_width / 2
```

##### Build intensity matrix for each bin in each scan

```python
intensity_matrix = np.zeros((len(RT), len(bin_centers)))
for i, (mz_array, intensity_array) in enumerate(zip(all_mz, all_intensity)):
    inds = np.digitize(mz_array, bins) - 1
    for idx, inten in zip(inds, intensity_array):
        if 0 <= idx < len(bin_centers):
            intensity_matrix[i, idx] += inten
```

#### Mass spectrometry imaging, IMZML file read example

##### import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from pyimzml.ImzMLParser import ImzMLParser
from tqdm import tqdm 
```

##### Defines the error range for a specified mass-to-charge ratio (m/z)

```python
def calculate_ppm_range(mz, ppm):
    """Calculate the valid range for a given m/z value and ppm error"""
    tolerance = mz * ppm / 1e6
    mz_min = mz - tolerance
    mz_max = mz + tolerance
    return mz_min, mz_max
```

##### Define file reading and image drawing main functions.

```python
def search_mz_and_plot(imzml_file, target_mz, ppm, img_set):
    """Search for spectra in the IMzML file based on the specified m/z value and plot the imaging map"""
    parser = ImzMLParser(imzml_file)
    all_coords = parser.coordinates  # Get all coordinates
    
    matching_coordinates, matching_intensities = find_spectra_in_range(parser, target_mz, ppm)


   

    # If matching spectra are found
    if matching_coordinates:
        plot_imaging_map(matching_coordinates, matching_intensities, img_set,
                         title=f"m/z = {target_mz} Imaging (ppm = {ppm})", 
                         all_coords=all_coords)
    else:
        print(f"No spectra found within the m/z = {target_mz} ({ppm} ppm error) range.")
```

##### Define data reading helper functions

```python
def find_spectra_in_range(parser, target_mz, ppm):
    """Find all spectra within the error range"""
    mz_min, mz_max = calculate_ppm_range(target_mz, ppm)
    matching_coordinates = []
    matching_intensities = []

    for coord in tqdm(range(len(parser.coordinates))):  # Add progress bar
        mz, intensity = parser.getspectrum(coord)
        
        # Find points in the spectrum where m/z falls within the error range
        mask = (mz >= mz_min) & (mz <= mz_max)
        if np.any(mask):  # If there are matching m/z values
            matching_coordinates.append(parser.coordinates[coord])
            matching_intensities.append(np.sum(intensity[mask]))  # Sum the intensity at this location

    return matching_coordinates, matching_intensities
```

##### Define image drawing helper functions

```python
def plot_imaging_map(coordinates, intensities, img_set, title="m/z Imaging", all_coords=None):
    """Plot the mass spectrometry imaging map, ensuring fixed image size and filling missing areas with zero intensity"""
    coordinates = np.array(coordinates)
    intensities = np.array(intensities)

    X = coordinates[:, 0]
    Y = coordinates[:, 1]
    
    # If all coordinates are provided, use them to calculate the image range
    if all_coords is not None:
        all_coords = np.array(all_coords)
        x_min = int(np.min(all_coords[:, 0]))
        x_max = int(np.max(all_coords[:, 0]))
        y_min = int(np.min(all_coords[:, 1]))
        y_max = int(np.max(all_coords[:, 1]))
    else:
        # Otherwise, calculate the range using only the matched points
        x_min = int(np.min(coordinates[:, 0]))
        x_max = int(np.max(coordinates[:, 0]))
        y_min = int(np.min(coordinates[:, 1]))
        y_max = int(np.max(coordinates[:, 1]))

    # Create a 2D array with size (y_max - y_min + 1, x_max - x_min + 1), initialized to zero
    image = np.zeros((y_max - y_min + 1, x_max - x_min + 1))

    # Fill intensity values
    for coord, intensity in zip(coordinates, intensities):
        x = int(coord[0]) - x_min
        y = int(coord[1]) - y_min
        image[y, x] = intensity  # Note: y corresponds to rows, x corresponds to columns
    X_axis = np.arange(x_min, x_max + 1)  
    Y_axis = np.arange(y_min, y_max + 1)  
    img_set.append((image, X_axis, Y_axis))

    # Set image size
    plt.figure(figsize=(8, 6))

    # Plot the image
    plt.imshow(image, cmap='hot', origin='lower')
    plt.colorbar(label="Intensity")
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel("X Coordinate", fontsize=14, fontweight='bold')
    plt.ylabel("Y Coordinate", fontsize=14, fontweight='bold')
    plt.show()
```
