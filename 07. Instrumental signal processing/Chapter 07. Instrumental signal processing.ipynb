```python
import numpy as np
from pylab import plot, ylim

from cwpy.signal.smooth import boxcar, savgol_coef
from scipy.signal import savgol_filter
```

```python
N = 1000 
dt = 1.0 / 100.0 
A = 5.0
frequency = 0.5
omega = 2 * np.pi * frequency

noise = 3 * (np.random.rand(N) - 0.5)
x = np.linspace(0.0, N*dt, N)
y = A * np.sin(omega * x) + noise

plot(x, y)
ylim(-8, 8)
```

```python
ys = boxcar(y, 5)
plot(ys)
```

```python
print(savgol_coef(2, 2, 0))

ys = savgol_filter(y, 11, 2)
plot(x, ys)
```
